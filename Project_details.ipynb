**Topic:** Understanding and predicting ecosystem disruption impacting mammals through machine learning models

**Climatology:** The study of long-term weather patterns and climate change from local to global scales. Climatology is essential for the understanding of the impact of climate change on ecosystems and human societies. It determines the future climate expectations involving analyzing weather data and understanding the factors that influence climate.

**Weather**
* The atmospheric conditions in a specific place at a particular time, such as temperature, humidity, precipitation, wind speed, etc.
* Examples: Thunderstorm, or a snowy

**Climate**
* The long-term average of weather patterns in a region over decades
* Examples: The Sahara Desert has a hot, arid climate

**What role do weather and climate play in ecosystems?**
  Weather and climate influence the distribution and abundance of species, and the availability of resources. Some species are adapted to a warm climate, while others thrive in cold or dry areas.

Examples: Polar bears depend on cold climates and sea ice to hunt seals
Many mammals time their breeding and migration with seasonal climate patterns


**Ecology**
* The study of the interaction between living organisms and their environment, including both biotic (living) and abiotic (non-living) factors

**Predation:** One organism (predator) hunts and eats another (prey).
    Example: A lion eating a zebra.

**Competition: **bold text** **Organisms compete for the same resources (food, space, mates).
    Example: Trees competing for sunlight in a forest.

**Mutualism:** Both species benefit from the relationship.
    Example: Bees pollinating flowers while feeding on nectar.

**Commensalism:** One species benefits while the other is neither helped nor harmed.
    Example: Birds nesting in trees.

**Parasitism:** One organism benefits (parasite) while the other is harmed (host).
    Example: Ticks feeding on a mammal’s blood.


**Disruption**

* When something causes a disturbance or break in how an ecosystem normally works.
* A wildfire that destroys a forest/Building a city where animals used to live.
